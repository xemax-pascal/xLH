```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-3, 3)
```

```python
a = 2.0  # Steigung
b = 1.0  # Achsenabschnitt

# Funktionsdefinition
y1 = a * x + b
y2 = x ** 2
y3 = x ** 3
```

```python
plt.plot(x, y1, label='y = ax + b')
plt.plot(x, y2, label='y = x2')

plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Funktionen')
```
